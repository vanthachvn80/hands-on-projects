## một ví dụ về làm sns.countplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset representing the survival status of breast cancer patients
data = {
    'status': ['Alive', 'Alive', 'Deceased', 'Alive', 'Deceased', 'Deceased', 'Alive', 'Deceased', 'Deceased']
}

# Create a DataFrame from the sample data
import pandas as pd
df = pd.DataFrame(data)

# Create the count plot
ax = sns.countplot(x='status', data=df)

# Add labels to the bars
for p in ax.patches:
   ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

# Add a title and labels
plt.title('Distribution of Survival Status of Breast Cancer Patients')
plt.xlabel('Survival Status')
plt.ylabel('Count')

# Display the plot
plt.show()
```
